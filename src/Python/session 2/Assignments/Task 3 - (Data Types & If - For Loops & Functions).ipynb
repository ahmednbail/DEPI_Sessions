### <span style="color:orange"><u>**17. Replace the last value of each tuple in a list**</u></span>

### <span style="color:orange"><u>**Find the longest word in a list**<u></span>

```python
words = ["hi", "hello", "fantastic", "world"]

# TODO: find the longest word
word=iter(words)
longest=next(word)
for i in word:
    if len(i) > len(longest):
         longest=i
print(longest)
# Hint: Track the word with the largest len().
```

```python
words = ["hi", "hello", "fantastic", "world"]
longest=max(words,key=len)
print(longest)
```

### <span style="color:orange"><u>**Sum values in a dictionary**<u></span>

```python
data = {"a": 10, "b": 20, "c": 30}

# TODO: sum all dictionary values
total=sum(data.values())
print(total)
```

#### <span style="color:red">**solution**</span>

```python
data = {"a": 10, "b": 20, "c": 30}
#for dictionary i have a key and value so you can loop and find what you need (key or values
total = 0
for value in data.values():
    total += value

print(total)
```
