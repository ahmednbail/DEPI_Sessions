### <span style="color:orange"><u>**3. Write a Python program to removes multiple items from a given set**</u></span>

#### <span style = "color:red"><u> **Example**</u></span>

- <u>my_set:</u> {1, 2, 3, 4, 5}

- <u>items to be removed:</u> {1, 4}

- <u>Output:</u> {2, 3, 5}

```python
# Write your answer here
def remove_items(s, items):
    for item in items:
        if item in s:
            s.remove(item)
    return s

my_set = {1, 2, 3, 4, 5}
items_to_remove = {1, 4}

print(remove_items(my_set, items_to_remove))
```

### <span style="color:orange"><u>**11. create a function that calculates the median of a list of numbers (a median is the middle element in sorted list of numbers)**</u></span>

#### <span style = "color:red"><u> **Example**</u></span>

- <u>numbers:</u> [5, 2, 7, 3, 1]

- <u>Output:</u> 3

```python
# Write your answer here
S=[5,2,7,3,1]
def median(s):
    s.sort()
    n = len(s)
    if n == 0:
        return None
    if n % 2 == 1:
      return s[n // 2]
    return s[n//2 - 1 : n//2 + 1]
print(median(S))
```

### <span style="color:orange"><u>**12. create a function that calculates the mode of a list of number (mode is the most frequent element in a list)**</u></span>

#### <span style = "color:red"><u> **Example**</u></span>

- <u>1st:</u> [5, 2, 7, 3, 2, 1, 5]

- <u>Output:</u> 5

```python
# Write your answer here
lst=[5, 2, 7, 3, 2, 1, 5]
from collections import Counter

def most_frequent(lst):
    if not lst:
        return None
    return Counter(lst).most_common(1)[0]

print(most_frequent(lst))
```

### <span style="color:orange"><u>**13. ctake the mark from the user and convert it to grade according to the following guidelines:**</u></span>

<span style = "color:red">**1.**</span> 85 - 100 (A)

<span style = "color:red">**2.**</span> 75 - 85 (B)

<span style = "color:red">**3.**</span> 65 - 75 (C)

<span style = "color:red">**4.**</span> 50 - 65 (D)

<span style = "color:red">**5.**</span> <50 (F)


```python
# Write your answer here
def get_grade(mark):
    if mark < 0 or mark > 100:
        return "Invalid mark"

    if 85 <= mark <= 100:
        return "A"
    elif 75 <= mark < 85:
        return "B"
    elif 65 <= mark < 75:
        return "C"
    elif 50 <= mark < 65:
        return "D"
    else:
        return "F"


mark = float(input("Enter your mark: "))
print("Grade:", get_grade(mark))
```
