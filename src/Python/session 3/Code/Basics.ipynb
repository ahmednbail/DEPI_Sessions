```python
def sum(a,b):
    return a+b
def diff(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b

```

```python
def main():
    while True :
        print("/n Calculator")
        print("1.  sum")
        print("2.  diff")
        print("3.  mul")
        print("4.  div")

        choice = input("Enter your choice: ")
        if choice == "5":
            print("calc stoped")
            break
        a=float(input("Enter first number: "))
        b=float(input("Enter second number: "))
        if choice == "1" :
            print("result:",sum(a,b))
        elif choice == "2":
            print("result:",diff(a,b))
        elif choice == "3":
            print("result:",mul(a,b))
        else:
             print("result",div(a,b))


if __name__=="__main__":
    main()


```
