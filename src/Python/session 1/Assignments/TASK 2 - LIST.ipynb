### <span style="color:orange"><u>**Exercise 3: Remove Element**<u></span>

- Create a list of fruits
- Ask user for fruit name
- Remove it if exists
- Hint:
Use append()


```python
fruits = ['Apple', 'Banana', 'Orange']
f = input("Enter fruit name: ")
#code here
if f in fruits:
     fruits.remove(f)
else:
    fruits.append(f)
print(fruits)
```
