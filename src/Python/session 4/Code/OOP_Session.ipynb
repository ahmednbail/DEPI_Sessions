```python
class FreqWords:
    def __init__(self,words):
        self.words=words
        unique_words=set(words)
    def freq(self):
         unique_words=set(self.words)
         for word in  unique_words:
             print(f'{word}:{self.words.count(word)} ')

```

```python
words=['welcome','python','python','python']
freq=FreqWords(words)
freq.freq()
```

```python
import  random
class Generator:
    def __init__(self,length):
        self.length=length
    def Pass(self):
        password=''
        characters='abcdefghijASDHFIHJKljalgjlakgrigairjglakgalekrng'
        for _ in range(self.length):
            password+=random.choice(characters)
        return  password
```

```python
from abc import  ABC , abstractmethod

class Animal(ABC):
    @abstractmethod
    def move (self):
        pass

class Bird(Animal):
    def move(self):
        print(' move from Bird')

class  Cat(Animal):
    def move(self):
        print(' move from Cat')

```

```python
a=Bird()
a.move()
```

```python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary

    def set_salar(self,obj,new_salary):
        obj.salary=new_salary

class Manager(Employee):
    def __init__(self, name, salary):
        super().__init__(name, salary)
        self.name=name
        self._salary=salary
    def get_salary(self):
        return self._salary
```

```python
E1=Employee('ahmed',2000)
m1=Manager('omar',5000)
m1.set_salar(m1,6000)
print(m1.get_salary())
```

```python
import random
def login_sys():
    u='test'
    p='1111'
    while True:
        x=input('enter your username:')
        if x ==u:
            y=input('enter your password:')
            if y ==p:
                s=random.randrange(10000,100000)
                print('Verification code is:',s)
                while True:
                    l=int(input('Enter the verification code:'))
                    if l ==s:
                        print('Access granted')
                        break
                    else:
                        print('invalid verification code : please try again')
                break
            else:
                print('Incorrect password!')
                continue
        else:
            print('Incorrect username !')
            continue
```

```python
login_sys()
```

```python
from prettytable import PrettyTable
products= [{"name":"water","price":80.0,"quantity":1500},
          {"name":"soda","price":130.0,"quantity":1500},
          {"name":"chips","price":200,"quantity":1500},
]
```

```python
def display_table(products):
    table=PrettyTable()
    table.field_names=["Product Name","Price($)","Quantity"]
    for p in products:
        table.add_row([p["name"],f"{p["price"]:.2f}",p["quantity"]])
    print(table)
```

```python
display_table(products)
```
