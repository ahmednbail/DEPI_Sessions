```python
import numbers


class EmailProcessor:
    def __init__(self, email):
        # Store email as an object attribute
        self.email = email

    def is_valid(self):
        # Check if email contains exactly one '@' and at least one '.'
        return self.email.count('@') == 1 and '.' in self.email

    def get_username(self):
        if not self.is_valid():
            return "Invalid email"
        return self.email.split('@')[0]

    def get_domain(self):
        if not self.is_valid():
            return "Invalid email"
        return self.email.split('@')[1].split('.')[0]


# ===== Testing =====
email_obj = EmailProcessor("Amit_ml@gmail.com")

print("Valid:", email_obj.is_valid())
print("Username:", email_obj.get_username())
print("Domain:", email_obj.get_domain())
```

```python
class VowelDecoder:
    def __init__(self, text):
        self.text = text

    def decode(self):
        # Replace shifted vowels
        return self.text.replace('E', 'A').replace('U', 'O')


# ===== Testing =====
decoder = VowelDecoder("EPUVT")
decoded_word = decoder.decode()
print(decoded_word)
```

```python
class MessageDecoder:
    def __init__(self, first_word, encoded_word):
        self.first_word = first_word
        self.encoded_word = encoded_word

    def decode_message(self):
        decoded = self.encoded_word.replace('E', 'A').replace('U', 'O')
        return f"{self.first_word} {decoded}"


# ===== Testing =====
message = MessageDecoder("projectplay", "EPUVT")
print(message.decode_message())
```

```python
# 4-
class SmallestNumberFinder:
    def __init__(self, numbers):
        self.numbers = numbers

    def find_smallest(self):
        smallest = self.numbers[0]
        for num in self.numbers:
            if num < smallest:
                smallest = num

        return smallest

numbers = [10, 3, 25, 1, 7]
finder = SmallestNumberFinder(numbers)
print(finder.find_smallest())
```

```python
# 5-
class LargestNumberFinder:
    def __init__(self, numbers):
        self.numbers = numbers

    def find_largest(self):
        largest = self.numbers[0]
        for num in self.numbers:
            if num > largest:
                largest = num
        return largest


numbers = [10, 3, 25, 1, 7]
finder = LargestNumberFinder(numbers)
print(finder.find_largest())

```

```python
# 11---
class SetAnalyzer:
    def __init__(self, data):
        self.data = data

    def find_max_min(self):
        iterator = iter(self.data)
        maximum = next(iterator)
        minimum = maximum
        for value in iterator:
            if value > maximum:
                maximum = value
            if value < minimum:
                minimum = value
        return maximum, minimum


my_set = {10, 3, 25, 1, 7}

analyzer = SetAnalyzer(my_set)
maximum, minimum = analyzer.find_max_min()
print("Max:", maximum)
print("Min:", minimum)
```

```python
# 12---
class TupleIndexer:
    def __init__(self, data):
        self.data = data

    def find_index(self, value):
        index = 0
        for item in self.data:
            if item == value:
                return index
            index += 1
        return -1


my_tuple = (10, 20, 30, 40)

indexer = TupleIndexer(my_tuple)
print(indexer.find_index(20))
```

```python
# 13--
class TupleToDictConverter:
    def __init__(self, data):
        self.data = data

    def convert(self):
        result = {}
        for key, value in self.data:
            result[key] = value
        return result


my_tuple = (("a", 1), ("b", 2), ("c", 3))

converter = TupleToDictConverter(my_tuple)
print(converter.convert())
```

```python
# 14---
class PairSplitter:
    def __init__(self, pairs):
        self.pairs = pairs

    def split(self):
        first = []
        second = []
        for a, b in self.pairs:
            first.append(a)
            second.append(b)
        return first, second


pairs = [(1, 2), (3, 4), (5, 6)]

splitter = PairSplitter(pairs)
list1, list2 = splitter.split()
print(list1)
print(list2)
```

```python
# 17-
class TupleModifier:
    def __init__(self, data):
        self.data = data

    def replace_last_value(self, new_value):
        result = []
        for t in self.data:
            temp = [*t]
            temp[-1] = new_value
            result.append(tuple(temp))
        return result


data = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]

modifier = TupleModifier(data)
new_list = modifier.replace_last_value(100)
print(new_list)
```

```python
# longest Word

class WordFinder:
    def __init__(self, words):
        self.words = words

    def find_longest(self):
        word = iter(self.words)
        longest = next(word)
        for i in word:
            if len(i) > len(longest):
                longest = i
        return longest


words = ["hi", "hello", "fantastic", "world"]

finder = WordFinder(words)
longest = finder.find_longest()
print(longest)
```

```python
#merg to dic
class DictMerger:
    def __init__(self, dict1, dict2):
        self.dict1 = dict1
        self.dict2 = dict2

    def merge(self):
        result = self.dict1.copy()
        result.update(self.dict2)
        return result


d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}

merger = DictMerger(d1, d2)
result = merger.merge()
print(result)
```

```python
class ScoreCalculator:
    def __init__(self, scores):
        self.scores = scores

    def sum_all_scores(self):
        total = 0
        for student_data in self.scores.values():
            for score in student_data.values():
                total += score
        return total


scores = {
    "student1": {"math": 80, "science": 90},
    "student2": {"math": 75, "science": 85},
    "student3": {"math": 92, "science": 88},
}

calculator = ScoreCalculator(scores)
total = calculator.sum_all_scores()
print(total)
```

```python
class FibonacciCalculator:
    def __init__(self, n):
        self.n = n

    def calculate(self):
        def fib(k):
            if k == 0:
                return 0
            if k == 1:
                return 1
            return fib(k-1) + fib(k-2)

        result = []
        for i in range(self.n):
            result.append(fib(i))
        return result

print(FibonacciCalculator(10).calculate())
```

```python
class SetRemover:
    def __init__(self, my_set):
        self.my_set = my_set

    def remove_item(self, item):
        if item in self.my_set:
            self.my_set.remove(item)
        return self.my_set

    def remove_multiple(self, items):
        for item in items:
            if item in self.my_set:
                self.my_set.remove(item)
        return self.my_set

print(SetRemover({1, 2, 3, 4, 5}).remove_item(3))
```

```python

class MedianCalculator:
    def __init__(self, numbers):
        self.numbers = numbers

    def calculate(self):
        sorted_nums = sorted(self.numbers)
        n = len(sorted_nums)
        if n == 0:
            return None
        if n % 2 == 1:
            return sorted_nums[n // 2]
        return sorted_nums[n//2 - 1 : n//2 + 1]

print(MedianCalculator([5, 2, 7, 3, 1]).calculate())
```

```python

class ModeCalculator:
    def __init__(self, lst):
        self.lst = lst

    def calculate(self):
        from collections import Counter
        if not self.lst:
            return None
        return Counter(self.lst).most_common(1)[0]
print(ModeCalculator([5, 2, 7, 3, 2, 1, 5]).calculate())
```

```python

class GradeConverter:
    def __init__(self, mark):
        self.mark = mark

    def get_grade(self):
        if self.mark < 0 or self.mark > 100:
            return "Invalid mark"

        if 85 <= self.mark <= 100:
            return "A"
        elif 75 <= self.mark < 85:
            return "B"
        elif 65 <= self.mark < 75:
            return "C"
        elif 50 <= self.mark < 65:
            return "D"
        else:
            return "F"
print(GradeConverter(45).get_grade())
```
